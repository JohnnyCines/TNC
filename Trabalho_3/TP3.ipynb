# Gerar chave RSA

```python
def Gera_Chave(nbits = 512):
    p = random_prime(2^(nbits//2), 2^(nbits//2-2))
    q = random_prime(2^(nbits//2 +1), 2^(nbits//2 -1))
    while p == q:
        p = random_prime(2^(nbits//2), 2^(nbits//2-2))
        q = random_prime(2^(nbits//2 +1), 2^(nbits//2 -1))
    n = p*q
    m = (p-1)*(q-1)
    e = randint(2, m-1)
    while gcd(e, m) != 1:
        e = randint(2,m-1)
    d = power_mod(e, -1, m)
    PubKey = (n, e)
    Chave = (PubKey, d)
    return Chave
```

```python
def encrypt(x, PubKey):
    n, e = PubKey
    criptograma = power_mod(x, e, n)
    return criptograma
```

```python
def decrypt(y, chave):
    n, _ = chave[0]
    d = chave[1]
    texto = power_mod(y, d, n)
    return texto
```

```python
chave = Gera_Chave(2048)
chave
```

```python
m = 1234
```

```python
cript = encrypt(m,chave[0])
cript
```

```python
dec = decrypt(cript, chave)
dec
```

# Assinatura cega de Chaum

```python
PubK, PrivK=Gera_Chave(2048)
m=1234
```

```python
n,e=PubK
k=randint(1,n-1)
while gcd(k,n)!=1:
    k=randint(1,n-1)
print("m: ",m)
print("e: ",e)
print("n: ",n)
print("k: ",k)
print("PrivK: ",PrivK)
```

```python
blind = (m * power_mod(k,e,n)) % n
```

```python
sign = (power_mod(blind,PrivK,n)) % n
```

```python
s = (power_mod(k,-1,n)*sign) % n
```

```python
print("blind: ",blind)
print("sign: ", sign)
print("Assinatura cega: ", s)
```
